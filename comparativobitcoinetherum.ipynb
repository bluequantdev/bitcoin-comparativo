```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

btc = yf.download('BTC-USD', start='2022-01-01', end='2025-07-27', interval='1d')
eth = yf.download('ETH-USD', start='2022-01-01', end='2025-07-27', interval='1d')

btc_monthly = btc['Close'].resample('M').last()
eth_monthly = eth['Close'].resample('M').last()

btc_ret = btc_monthly.pct_change() * 100
eth_ret = eth_monthly.pct_change() * 100

df = pd.concat([btc_ret, eth_ret], axis=1)
df.columns = ['BTC %', 'ETH %']
df = df.dropna()

plt.figure(figsize=(12,6))
plt.plot(df.index, df['BTC %'], marker='o', label='Bitcoin (BTC)')
plt.plot(df.index, df['ETH %'], marker='s', label='Ethereum (ETH)')
plt.axhline(0, color='gray', linestyle='--', alpha=0.5)
plt.title('Comparativo de Ganho Percentual Mensal: Bitcoin vs Ethereum')
plt.xlabel('Mês')
plt.ylabel('Ganho Percentual (%)')
plt.legend()
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()

```
